```python
from torchvision import datasets
from torchvision.transforms import Compose, AutoAugment, ToTensor
from pathlib import Path
```

```python
data_path = "../data-unversioned/p1ch7/"
download = not Path(data_path).exists()

custom_transforms = Compose([
    AutoAugment(),
    ToTensor()
])

cifar10 = datasets.CIFAR10(data_path, train=True, transform=custom_transforms, download=download)
cifar10_val = datasets.CIFAR10(data_path, train=False, download=download)
```
